```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Loading the image
img = cv2.imread('drive/MyDrive/images/assign-1/0_mask_2.jpg')
hsv = cv2.cvtColor(img , cv2.COLOR_BGR2HSV)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
# Define range of green color in HSV
Green_low = np.array([50, 155, 80])
Green_high = np.array([60, 255, 255])

# Threshold the HSV image to get only the red colors
mask_green = cv2.inRange(hsv,Green_low,Green_high)

plt.imshow(mask_green, cmap="gray") #cv2_imshow(mask_green)
```

```python
# Define range of red color in HSV
Red_low = np.array([0, 155, 80])
Red_high = np.array([9, 255, 255])

# Threshold the HSV image to get only the red colors
mask_red = cv2.inRange(hsv,Red_low,Red_high)

plt.imshow(mask_red, cmap="gray") #cv2_imshow(mask_red)
```

```python
# Define range of light_blue color in HSV
LBlue_low = np.array([85, 155, 80])
LBlue_high = np.array([95, 255, 255])

# Threshold the HSV image to get only the light_blue colors
mask_lblue = cv2.inRange(hsv,LBlue_low,LBlue_high)

plt.imshow(mask_lblue, cmap="gray") #cv2_imshow(mask_lblue)
```

```python
# Define range of yellow color in HSV
Yellow_low = np.array([25, 155, 80])
Yellow_high = np.array([35, 255, 255])

# Threshold the HSV image to get only the yellow colors
mask_yellow = cv2.inRange(hsv,Yellow_low,Yellow_high)

plt.imshow(mask_yellow, cmap="gray") #cv2_imshow(mask_yellow)
```

```python
#Processing All Masks
# Converting those pixels with values 1-127 to 0 and others to 1
threshold_green = cv2.threshold(mask_green, 127, 255, cv2.THRESH_BINARY)[1]
threshold_red = cv2.threshold(mask_red, 127, 255, cv2.THRESH_BINARY)[1]
threshold_lblue = cv2.threshold(mask_lblue, 127, 255, cv2.THRESH_BINARY)[1]
threshold_yellow = cv2.threshold(mask_yellow, 127, 255, cv2.THRESH_BINARY)[1]

plt.imshow(threshold_green, cmap="gray")
```

```python
# Apply the Component analysis function
analysis_green = cv2.connectedComponentsWithStats(threshold_green,
                                            4,
                                            cv2.CV_32S)
(totalLabels_g, label_ids, values_g, centroid) = analysis_green
```

```python
sum_g=0
# Loop through each component
for i in range(1, totalLabels_g):
   
      # Area of the component
    area = values_g[i, cv2.CC_STAT_AREA]
    sum_g = sum_g+area
    #print(area)
```
